# Explication du Dataset : Credit Card Customer Segmentation

Le dataset **Credit Card Customer Segmentation Dataset** contient des informations sur le comportement d'utilisation des cartes de crédit de clients pendant environ **6 mois**. Chaque ligne représente un client.

Voici l'explication claire de chaque colonne :

---

## 1️⃣ Informations d'identification

### `CUST_ID`

- Identifiant unique du client
- Sert seulement à identifier chaque client
- ❗ **On ne l'utilise pas dans le modèle K-Means**

**Exemple :**

```
C10001
C10002
```

---

## 2️⃣ Solde et utilisation de la carte

### `BALANCE`

- Solde moyen du compte du client pendant la période étudiée

**Exemple :**  
Si un client garde 1500$ sur sa carte, alors :

```
BALANCE = 1500
```

### `BALANCE_FREQUENCY`

- Fréquence de mise à jour du solde (entre 0 et 1)

| Valeur | Signification                       |
| ------ | ----------------------------------- |
| 1      | solde mis à jour très régulièrement |
| 0      | presque jamais                      |

---

## 3️⃣ Comportement d'achat

### `PURCHASES`

- Montant total des achats effectués avec la carte

**Exemple :**

```
PURCHASES = 3200
```

### `ONEOFF_PURCHASES`

- Achats effectués en une seule fois

**Exemple :**

- acheter une TV → one-off purchase

```
ONEOFF_PURCHASES = 1500
```

### `INSTALLMENTS_PURCHASES`

- Achats payés en plusieurs fois

**Exemple :**

- acheter un téléphone payé en 12 mensualités

```
INSTALLMENTS_PURCHASES = 800
```

### `PURCHASES_FREQUENCY`

- Fréquence des achats (0 → 1)

| Valeur | Signification |
| ------ | ------------- |
| 0      | très rarement |
| 1      | très souvent  |

### `ONEOFF_PURCHASES_FREQUENCY`

- Fréquence des achats en une seule transaction

### `PURCHASES_INSTALLMENTS_FREQUENCY`

- Fréquence des achats en plusieurs paiements

---

## 4️⃣ Avances de cash

### `CASH_ADVANCE`

- Montant total retiré en cash avec la carte de crédit

**Exemple :**

- retirer 500$ avec la carte

```
CASH_ADVANCE = 500
```

### `CASH_ADVANCE_FREQUENCY`

- Fréquence des retraits cash

### `CASH_ADVANCE_TRX`

- Nombre de transactions de retrait cash

**Exemple :**

```
CASH_ADVANCE_TRX = 5
```

→ le client a retiré du cash **5 fois**

---

## 5️⃣ Activité globale

### `PURCHASES_TRX`

- Nombre total de transactions d'achat

**Exemple :**

```
PURCHASES_TRX = 20
```

→ **20 achats** effectués

---

## 6️⃣ Limite de crédit

### `CREDIT_LIMIT`

- Montant maximum que la banque autorise au client

**Exemple :**

```
CREDIT_LIMIT = 10000
```

→ le client peut utiliser jusqu'à **10 000$**

---

## 7️⃣ Paiements

### `PAYMENTS`

- Montant total que le client a payé pour rembourser sa carte

**Exemple :**

```
PAYMENTS = 3000
```

### `MINIMUM_PAYMENTS`

- Montant minimum que le client doit payer chaque mois

---

## 8️⃣ Utilisation de la carte

### `PRC_FULL_PAYMENT`

- Pourcentage de fois où le client a payé la totalité du solde

| Valeur | Signification       |
| ------ | ------------------- |
| 1      | paie toujours tout  |
| 0      | ne paie jamais tout |

---

## 9️⃣ Durée de la relation client

### `TENURE`

- Nombre de mois depuis que le client utilise la carte

**Exemple :**

```
TENURE = 12
```

→ client depuis **12 mois**
