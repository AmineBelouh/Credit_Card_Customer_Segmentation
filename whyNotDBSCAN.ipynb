## 1️⃣ Peut-on appliquer DBSCAN sur ce dataset ?

**Oui**, c'est techniquement faisable, mais ce n'est **pas l'algorithme le plus adapté** pour ce dataset.

Le dataset **Credit Card Customers Dataset** contient des informations sur le comportement d'utilisation des cartes de crédit :

- `BALANCE`
- `PURCHASES`
- `CASH_ADVANCE`
- `PAYMENTS`
- `PURCHASES_FREQUENCY`
- `CREDIT_LIMIT`
- etc.

### Le but du projet est généralement :

➡️ **Segmenter les clients selon leur comportement financier**

Donc nous cherchons :

- groupes de clients similaires
- profils marketing
- stratégies commerciales

Pour ce type d'objectif, le **clustering par centroïdes** est très efficace, d'où l'utilisation de **K-Means clustering**.

---

## 2️⃣ Pourquoi nous avons utilisé K-Means

K-Means est adapté parce que :

### ✔ Objectif : segmentation clients

K-Means permet de créer :

- clients actifs
- clients faibles utilisateurs
- clients cash advance
- clients premium

C'est exactement ce qu'on veut pour :

- marketing
- gestion du risque
- offres personnalisées

### ✔ Les clusters sont globaux

Dans ce dataset, les clients forment des groupes relativement homogènes.

**Exemple :**

| Client type               | Comportement         |
| ------------------------- | -------------------- |
| gros utilisateurs         | beaucoup d'achats    |
| utilisateurs modérés      | achats moyens        |
| utilisateurs cash advance | beaucoup de retraits |

**K-Means fonctionne très bien pour ce type de structure.**

---

## 3️⃣ Pourquoi DBSCAN n'est pas idéal ici

L'algorithme **DBSCAN** fonctionne différemment.

Son objectif principal est :

➡️ **Détecter les points isolés (anomalies)**

Il est souvent utilisé pour :

- fraude bancaire
- détection d'anomalies
- détection de bruit dans les données

### Problème avec ce dataset

Le dataset des cartes de crédit :

- contient beaucoup de variables
- contient beaucoup de points
- ne contient pas forcément des anomalies très isolées

**DBSCAN peut donc :**

- classer beaucoup de points comme bruit
- créer des clusters instables
- être très sensible aux paramètres `eps` (rayon de voisinage) et `min_samples`(nombre minimum de voisins)

---

## 4️⃣ Autre problème : dimension élevée

Ce dataset contient environ **17 variables**.

Or DBSCAN fonctionne moins bien quand :

```
dimension élevée
```

Parce que :

```
les distances deviennent moins significatives
```

C'est un problème classique appelé :

➡️ **Malédiction de la dimension**

---

## 5️⃣ Comparaison simple

| Critère                 | K-Means      | DBSCAN              |
| ----------------------- | ------------ | ------------------- |
| Objectif                | segmentation | détection anomalies |
| Besoin nombre clusters  | oui          | non                 |
| Adapté au dataset       | ✔ oui        | ❌ moins            |
| Interprétation business | facile       | difficile           |

---

## 6️⃣ Conclusion

### Pourquoi nous avons choisi K-Means plutôt que DBSCAN

Même si l'algorithme DBSCAN peut être appliqué à ce dataset, il n'est pas le plus adapté pour notre objectif.

Notre but est de **segmenter les clients selon leur comportement d'utilisation de la carte de crédit** afin d'identifier différents profils clients.

L'algorithme **K-Means** est particulièrement efficace pour ce type de problème car il permet de regrouper les clients ayant des comportements similaires en **clusters facilement interprétables** pour des décisions marketing.

À l'inverse, **DBSCAN** est principalement conçu pour **détecter des anomalies ou des points isolés** dans les données. Or notre dataset ne vise pas à détecter des fraudes ou des anomalies mais plutôt à identifier des groupes homogènes de clients.

De plus, DBSCAN peut être **moins performant sur des datasets de dimension élevée**, comme c'est le cas ici avec plusieurs variables financières, et il est très sensible au choix des paramètres.

Pour ces raisons, **K-Means est plus adapté à notre problématique de segmentation client**, tandis que DBSCAN serait plus pertinent dans un contexte de détection de fraude bancaire.
